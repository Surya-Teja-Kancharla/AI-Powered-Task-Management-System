```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    words = [word for word in words if word.isalnum() and word not in stop_words]
    ps = PorterStemmer()
    stemmed = [ps.stem(word) for word in words]
    return ' '.join(stemmed)
```

```python
import pandas as pd

data = pd.read_csv('New_Task_data.csv')
```

```python
data['Processed_Description'] = data['Task_Description'].apply(preprocess_text)
data['Processed_Shared_Taskboard_Updates'] = data['Shared_Taskboard_Updates'].apply(preprocess_text)
data.to_csv('preprocessed_task_data.csv', index=False)
```
