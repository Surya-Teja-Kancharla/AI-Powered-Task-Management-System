```python
import pandas as pd

data = pd.read_csv("preprocessed_task_data.csv")
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine the two columns into a single column
data['Combined_Descriptions'] = data['Processed_Description'] + " " + data['Processed_Shared_Taskboard_Updates']

# Apply TfidfVectorizer on the combined column
vectorizer = TfidfVectorizer(max_features=500)
features = vectorizer.fit_transform(data['Combined_Descriptions']).toarray()
```

```python
import numpy as np
np.save('task_features.npy', features)
```
